```python
guide = pd.read_excel('C:/Users/1645295/Desktop/Geek/Итоговая аттестация/Премиум авто.xlsx', 'Справочник')
guide
```

```python
marketing_data.loc[marketing_data['Domain'].str.contains('bmw', case=False, regex=True), 'Марка'] = 'BMW'
marketing_data.loc[marketing_data['Domain'].str.contains('mersedes', case=False, regex=True), 'Марка'] = 'Mersedes'
marketing_data
```

```python
models = list(map(str, guide['Модель'])) 
models
```

```python
for i in range(len(models)):
    marketing_data.loc[marketing_data['Goal Completion Location'].str.contains(models[i], case=False, regex=True), 'Модель'] = models[i]
marketing_data
```

```python
with pd.ExcelWriter("Премиум_авто_fixed_data.xlsx") as writer:
   

    marketing_data.to_excel(writer, sheet_name="Маркетинговые данные", index=False)
    guide.to_excel(writer, sheet_name="Справочник", index=False)
    data_CRM.to_excel(writer, sheet_name="Данные из CRM", index=False)
```
